```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#defining the main CSV file
covid = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_19-covid-Confirmed.csv&filename=time_series_2019-ncov-Confirmed.csv'
```

```python
#Showing the first five columns
d1=pd.read_csv(covid)
d1.head()
```

```python
#accesing the index of the column by adding a new row named 'Zero Cases' 
d1.loc ['Zero Cases'] = d1.isin([0]).sum()
d1
```

```python
#dropping the columns which are having 70% number of confirmed COVID-19 cases as zero [70% of 488 rows= (341 approx.)]
d1 = d1.loc[:,d1.loc['Zero Cases'] <= 341]
d1 = d1.drop(['Zero Cases'], axis = 0)
d1
```

```python
#Using the columns to plot the graphs Country wise (Countries vs number of cases)
df1 = d1.groupby(['Country/Region']).mean()
df1_ = df1.iloc[2:,2:]
df1_['Mean'] = df1_.mean(axis = 1)
df1_ = df1_[df1_['Mean']>200]
df1_ = df1_.drop(['Mean'], axis = 1)
df1_1 = df1_.iloc[:,-1]
df1_1.plot(kind = 'barh', figsize = (15,7), color='red')
plt.title('COVID-19 cases (as on 22 Mar 2020)')
plt.show()
```

```python
#Using the columns to plot the graphs Date-wise (Date vs number of cases) & World-Wide 
df2 = df1_.transpose()
df2 = df2.iloc[2:,1:]
```

```python
df2.plot(figsize = (15,7))
plt.legend(loc = 'best')
plt.title('COVID 19 cases  by Country')
plt.show()
```

```python
#Plotting the no of cases of a particular Country(Italy) in a pie chart 
df3 = df2['Italy']
df3.plot(kind = 'pie', figsize = (15,5))
plt.title('COVID 19 cases - Italy')
plt.show()
```

```python
#Plotting the no of cases of a particular Country(China) in a pie chart 
df3 = df2['China']
df3.plot(kind = 'pie', figsize = (15,5))
plt.title('COVID 19 cases - China')
plt.show()
```

```python
#Plotting the no of cases of a particular Country(Iran) in a pie chart 
df3 = df2['Iran']
df3.plot(kind = 'pie', figsize = (15,5))
plt.title('COVID 19 cases - Iran')
plt.show()
```

```python
#now we use the edited file for continent-wise graph (first download the CSV file or the dataset, then add the column 'continent'
# & after that fill up that column by adding the continents according to the respective countries)  
dfc = pd.read_csv('time_series_2019-ncov-Continent.csv')
```

```python
#Using the columns to plot the graphs Continent-wise (Continent vs number of cases)
dfc = dfc.groupby(['Continent']).sum()
dfc = dfc.iloc[:,-1]
dfc.plot(kind = 'barh', figsize = (15,7), color='red')
plt.title('COVID 19 cases by Continent as on 22 Mar 2020')
plt.show()
```
